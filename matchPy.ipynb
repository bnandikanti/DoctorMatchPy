```python
match_df = pd.read_csv("./match_file.csv")
match_df = match_df.dropna(how = 'any')
match_df.head()
```

```python
# # of Doctors matched with NPI
counter = 0
NPIMatchCounter = 0
with open('source_data.json') as data_file:    
    data = json.load(data_file)
for index, row in match_df.iterrows():
    for c in data:
        counter += 1
        if row['npi'] == c["doctor"]["npi"]:
            print(c["doctor"])
            NPIMatchCounter += 1
print(f"Total number of documents scanned : {counter}")
print(f"Total number of doctors matchec : {NPIMatchCounter}")
print(f"Number not matched : {counter - NPIMatchCounter}")
```

```python
# # of Practice matched with address
practiceMatchCounter = 0
with open('source_data.json') as data_file:    
    data = json.load(data_file)
for index, row in match_df.iterrows():
    for value in data:
        for i in range(len(value['practices'])): 
             if row['state'].lower() == value["practices"][i]["state"].lower() and \
                row['city'].lower() == value["practices"][i]["city"].lower() and \
                row['street'].lower() == value["practices"][i]["street"].lower():
                    practiceMatchCounter += 1
                    print(value["practices"]) 
print(practiceMatchCounter)
```

```python
# # of Doctors matched with name and address
nameAndPracticeMatchCounter = 0
with open('source_data.json') as data_file:    
    data = json.load(data_file)
for index, row in match_df.iterrows():
    for value in data:
        for i in range(len(value['practices'])): 
             if row['first_name'].lower() == value["doctor"]["first_name"].lower() and \
                row['last_name'].lower() == value["doctor"]["last_name"].lower() and \
                row['state'].lower() == value["practices"][i]["state"].lower() and \
                row['state'].lower() == value["practices"][i]["state"].lower() and \
                row['city'].lower() == value["practices"][i]["city"].lower() and \
                row['street'].lower() == value["practices"][i]["street"].lower():
                    nameAndPracticeMatchCounter += 1
                    print(value["doctor"]) 
print(nameAndPracticeMatchCounter)

```
